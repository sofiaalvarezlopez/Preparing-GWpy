<h1 align="center">Understanding filters</h1>

```python
from gwpy.signal.filter_design import notch, bandpass, highpass, lowpass
import matplotlib.pyplot as plt
from gwpy.plot import BodePlot, Plot
```

```python
# Notch filter 
n = notch(100, 4096)
plot = BodePlot(n, sample_rate=4096, c='limegreen')
ax = plot.gca()
ax.grid(0)
plot.suptitle("Notch filter")
```

```python
# Low pass
lp = lowpass(1000, 4096)
plot = BodePlot(lp, sample_rate=4096, c='limegreen')
ax = plot.gca()
ax.grid(0)
plot.suptitle("Lowpass filter")
```

```python
hp = highpass(100, 4096)
plot = BodePlot(hp, sample_rate=4096, c='limegreen')
ax = plot.gca()
ax.grid(0)
plot.suptitle("Highpass filter")
```

```python
bp = bandpass(100, 1000, 4096)
plot = BodePlot(bp, sample_rate=4096, c='limegreen')
ax = plot.gca()
ax.grid(0)
plot.suptitle("Bandpass filter")
```
